```python
#Checked in Tableau, Counted Distinct customer_id column to check if all values are unique,
#hence instead of dropping it we will be using it as a unique identifier.
df = df.rename(columns={'Unnamed: 0':'customer_id'})
```

```python
#Dataset seems mostly clean but we are going to do a sanity check and drop any duplicates if any.
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.dtypes
```

```python
#Lowercasing the columns as well as adding '_' to make it organized 
df.columns = df.columns.str.strip()\
                       .str.lower()\
                       .str.replace(' ','_')
```

```python
#Converting Churn column from 'int' to 'bool'

df['churn'] = df['churn'].astype(bool)
```

```python
df.to_csv('df_clean.csv')
```
